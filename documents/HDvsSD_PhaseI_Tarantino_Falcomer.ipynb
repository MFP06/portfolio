# Exercice 5 : 


```python
import soundfile as sf
from pylab import * 

def veecteurs(files):
    ve, fs = sf.read(files)
    with sf.SoundFile(files) as v:
        if v.channels == 1:
            droite = ve
            gauche = ve
        else:
            droite = ve[:, 0]
            gauche = ve[:, 1]
    return droite, gauche, fs

def plotFreq(signal, fe, couleur, titre):
    N = len(signal)  
    spectre = abs((1 / N) * fftshift(fft(signal)))  
    freq = arange(-fe / 2, fe / 2, fe / N) 

    plt.plot(freq, spectre, couleur + '-')  
    plt.grid()  
    plt.title(titre, fontsize=16)  
    plt.xlim([-fe / 2, fe / 2]) 
    plt.xlabel('Fréquence (Hz)', fontsize=16) 
    plt.ylabel('Amplitude', fontsize=16)  

 

def doublepiste(nom_fichier):
    canalG, canalD, fs = veecteurs(nom_fichier)  # ← récupère fs

    plt.figure(figsize=(20, 8))

    plt.subplot(121)
    plotFreq(canalG, fs, 'b', 'Piste gauche')

    plt.subplot(122)
    plotFreq(canalD, fs, 'r', 'Piste droite')

    plt.tight_layout()
    plt.show()


print(doublepiste('AJS_short.flac'))
print(doublepiste('Hoo_short.flac'))
print(doublepiste('PF_short.flac'))
print(doublepiste('DS_short.wav'))
```

# Exercice 8 :

```python

import soundfile as sf
from pylab import * 

def intro(files):
    ve, fs = sf.read(files)
    intro = ve[:(3* fs)]
    return intro

def newfiles(files):
    f = read(files)[1]
    intro2 = intro(files) 
    intro_fichier = files.replace("_short.flac", ".flac")
    write(intro_fichier, intro2, f)
    
newfiles('AJS_short.flac')
newfiles('Hoo_short.flac')
newfiles('PF_short.flac')
va, fa = read('AJS.flac')
vh, fh = read('Hoo.flac')
vp, fp = read('PF.flac')
la = len(va)
lh = len(vh)
lp = len(vp)
dureea = la / fa
dureeh = lh / fh
dureep = lp / fp

print('La durée de AJS.flac est de', dureea, 'sec')
print('La durée de Hoo.flac est de', dureeh, 'sec')
print('La durée de PF.flac est de', dureep, 'sec')
```
